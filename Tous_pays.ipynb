```python
## Read html and parse cells line by line 
with open ('my_iso_code.txt', 'r') as fp:
    html = fp.read()
    soup = BeautifulSoup(html, 'html5lib')
```

```python
#fonction qui cherche un valeur sur WB en fonction du code du pays donné et de l'indicateur de colonne 
def get_indicator(country_code, indicator):
    head ='http://api.worldbank.org/v2/country/'
    middle = '/indicator/'
    tail = '?format=json'
    url = head+country_code+middle+indicator+tail
    r= requests.get(url)
    try :
        data = r.json()
        if data[1] is None:
            return None
        else:
            
            for i in data[1]:
                if i['value'] is None:
                    continue
                else : 
                    return i['value']
    except KeyError:
        return None
    except IndexError:
        return None

#TEST
print(get_indicator('FRA','GC.TAX.IMPT.ZS'))
```

```python
#fonction qui rempli toutes les lignes d'une colonne du tableau, pour un indicateur donné on resort toutes les valeurs pour tous les pays dont on a le code 
def remplir_col(indic, nom_col):      
    for i,pays in df1['Country Name'].iteritems():
        country_code = countries_dic[pays]
        value = get_indicator(country_code, indic)
        print(value)
        if value is not None :
            df1.at[i, nom_col]=value
        print(pays, 'done...')
        time.sleep(0.5)
    df1[['Country Name',nom_col]]
```

```python
#test pour remplir une seule colonne 
remplir_col('FB.AST.NPER.ZS', 'WB IMF Non-performing loans to total gross loans ratio (%)')
```

```python
#Remplissage de tout le tableau colonne après colonne 
def remplir_tab(list_indic) :
    for indicator in list_indic :
        remplir_col(list_indic[indicator], indicator)
```
