```python
## Import useful libraries
from bs4 import BeautifulSoup
import pickle

## Read html and parse cells line by line 
with open ('my_iso_code.txt', 'r') as fp:
    html = fp.read()
    soup = BeautifulSoup(html, 'html5lib')

```

```python
#regarder documentation beautifulSoup
trs=soup.find_all('tr')
len(trs)


for tr in trs:
    values=[td.string for td in tr.find_all('td')]
    if values[0] is not None and len(values[0])>0 :
        print(values[0], values[3])


```

```python
#2e posibilité
countries=[]
for tr in trs: 
    values=[td.string for td in tr.find_all('td')]
    countries.append((str(values[0]), str(values[3])))
countries=countries[1: -1]
```

```python
countries
```

```python
import pickle
with open ('all_countries.pickle', 'wb+') as fp:
    pickle.dump(countries, fp)
```
